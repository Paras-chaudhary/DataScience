```python
#DATA VISUALISATION
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = x**2 + 3
print(x,y)
```

```python
plt.plot(x,y,label ="Stock Price")
plt.plot(x,x**2,label ="Stock")
plt.legend()
plt.xlabel("timestamp")
plt.ylabel("price")
plt.title("Stock")
plt.style.use("seaborn")
```

```python
plt.bar(x,y,label ="Stock Price")
plt.bar(x,x**2,label ="Stock")
plt.legend()
plt.xlabel("timestamp")
plt.ylabel("price")
plt.title("Stock")
plt.style.use("seaborn")
```

```python
plt.scatter(x,y,label ="Stock Price")
plt.scatter(x,x**2,label ="Stock")
plt.legend()
plt.xlabel("timestamp")
plt.ylabel("price")
plt.title("Stock")
plt.style.use("seaborn")
```
